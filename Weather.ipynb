```python
weather = {}

for i in range(2,9):
    url = "http://api.apixu.com/v1/history.json?key=3d155e6fb2b3462583b90831190907&q=rg1&dt=2019-07-"+str(i)
    response = requests.get(url)
    data = response.json()
    day = data["forecast"]["forecastday"][0]["hour"]
    days = {}
    for j in range(0,24):
        days[j] = day[j]["condition"]["text"]
    weather["2019-07-"+str(i)] = days
```

```python
display(weather["2019-07-2"])
```

```python
df_weather = pd.DataFrame(weather)
display(df_weather)
```

```python
timings = {}

for i in range(2,9):
    url = "https://rtl2.ods-live.co.uk/api/trackingHistory?key=hvOtkiqAwK&service=17&date=2019-07-"+str(i)+"&vehicle=707&location="
    response = requests.get(url)
    data = response.json()
    day = []
    for track in data:
        times = [track["ScheduledArrivalTime"], track["ScheduledDepartureTime"], 
               track["ArrivalTime"], track["DepartureTime"]]
        day.append(times)
    timings["2019-07-"+str(i)] = day
```

```python
display(timings)
```

```python
mapping = {}
for date in timings.keys():
    times_weather = []
    for track in timings[date]:
        if track[2] != "":
            arrival_time = time.strptime(track[2], "%Y-%m-%d %H:%M:%S")
            weather_cond = weather[date][arrival_time.tm_hour]
            track.append(weather_cond)
            times_weather.append(track)
    mapping[date] = times_weather
```

```python
mapping["2019-07-8"][1]
```

```python
weather_punct = []
for day in mapping:
    for stop in mapping[day]:
        scheduled_arrival = time.strptime(stop[0], "%Y-%m-%d %H:%M:%S")
        real_arrival = time.strptime(stop[2], "%Y-%m-%d %H:%M:%S")
        weather_punct.append(((time.mktime(real_arrival)-time.mktime(scheduled_arrival)), stop[4]))
```

```python
weather_punct[1]
```

```python
df = pd.DataFrame(weather_punct, columns=["punctuality", "weather"])
df
```

```python
weather_names = df['weather'].unique().tolist()
weather_names
```
